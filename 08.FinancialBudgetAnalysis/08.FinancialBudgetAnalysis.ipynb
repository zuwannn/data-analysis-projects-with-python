### Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

dataset contains data about the financial budget of India for the year 2021

```python
data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/India_budget_2021.csv")
data.head()
```

```python
print(data)
```

```python
data.dropna()
```

 prepare the data by only selecting the main departments and putting all the other departments in the other category

```python
data = data.iloc[[0,8,11,14,18,23,41,42,43],:]
row = {'Department /Ministry': 'OTHERS', 'Fund allotted(in ₹crores)': 592971.0800000001}
data = data.append(row, ignore_index = True)
print(data)
```

plot this data to have a look at the priorities of the government for the financial year

```python
data.plot.bar(x='Department /Ministry', y='Fund allotted(in ₹crores)')
```

```python
df = data["Fund allotted(in ₹crores)"]
labels = data["Department /Ministry"]
plt.figure(figsize=(7,7))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85, shadow=True)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Department wise Budget", fontsize=20)
plt.show()
```
